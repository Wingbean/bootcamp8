```python
#!pip install gazpacho
from gazpacho import Soup
import requests

url = "https://www.rottentomatoes.com/browse/movies_at_home/affiliates:netflix~sort:popular"
html = requests.get(url)
soup = Soup(html.text)
```

```python
titles = soup.find("span",{"class": "p--small"})
#len(titles)
#titles[].strip
clean_titles = []
for title in titles:
    clean_titles.append(title.strip())

del clean_titles[0:4]

clean_titles
```

```python
dates = soup.find("span",{"class": "smaller"})

clean_dates = []
for date in dates:
    clean_dates.append(date.strip())

clean_dates
```

```python
import pandas as pd
tomatoes = pd.DataFrame(data= {
    "title": clean_titles,
    "date": clean_dates
})

tomatoes
```

```python
percents = soup.find("span",{"class": "percentage"})
percents
```

```python
clean_percents = []
for percent in percents:
    clean_percents.append(percent.strip())

clean_percents
```
