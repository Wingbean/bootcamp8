```python
# read country file
import pandas as pd

url = "https://pkgstore.datahub.io/core/country-list/data_csv/data/d7c9d7cfb42cb69f4422dec222dbbaa8/data_csv.csv"
coun = pd.read_csv(url)
coun_names = list(coun["Name"])

# API
import requests
import time

countries = []
regions = []
pops = []
areas = []
for i in coun_names:
    url = f"https://restcountries.com/v3.1/name/{i}"
    resp = requests.get(url)
    if resp.status_code == 200:
        json_data = resp.json()
        countries.append(json_data[0]["name"]["common"])
        regions.append(json_data[0]["region"])
        pops.append(json_data[0]["population"])
        areas.append(json_data[0]["area"])
    else:
        countries.append("error")
        regions.append("error")
        pops.append("error")
        areas.append("error")
    #time.sleep(1)

myproj = pd.DataFrame(data= {
    "Country": countries,
    "Region": regions,
    "Area": areas,
    "population": pops
})

myproj
```
